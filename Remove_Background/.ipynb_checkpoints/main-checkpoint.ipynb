```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("input.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_hsv)
```

```python
lower_kiwi = np.array([11, 68, 0]) 
upper_kiwi = np.array([41, 255, 236])
```

```python
mask = cv2.inRange(img_hsv, lower_kiwi, upper_kiwi)
```
