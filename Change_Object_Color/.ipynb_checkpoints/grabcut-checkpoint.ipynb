```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('input.jpg')
```

```python
plt.imshow(img)
```

```python
height, width = img.shape[:2]
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
bgModel = np.zeros((1, 65), np.float64)
fgModel = np.zeros((1, 65), np.float64)
```

```python
rect = (50, 50, width-100, height-100)
```

```python
cv2.grabCut(img, mask, rect, bgModel, fgModel, 5, cv2.GC_INIT_WITH_RECT)
```

```python
mask_2 = np.where((mask == 2) | (mask == 0), 0, 1).astype('uint8')
grabcut_result = img * mask_2[:, :, np.newaxis]
```

```python
kernel = np.ones((5, 5), np.uint8)
cleaned_mask = cv2.morphologyEx(mask_2, cv2.MORPH_OPEN, kernel)
```

```python
final_result = img * cleaned_mask[:, :, np.newaxis]
```

```python
plt.imshow(final_result)
```

```python
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
```

```python
mask_all = cv2.bitwise_or(mask1,mask2)
```

```python
mask_all = cv2.dilate(mask, np.ones((9,9), np.uint8), iterations=5)
mask_all = cv2.GaussianBlur(mask, (3,3), 0)
```

```python
hsv[:, :, 0][mask_all > 0] = 140
```

```python
purple_rose = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
```

```python
purple_rose = purple_rose * cleaned_mask[:, :, np.newaxis]
```
