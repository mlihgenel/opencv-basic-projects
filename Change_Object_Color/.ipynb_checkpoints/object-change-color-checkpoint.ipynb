```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input.jpg')
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
def display_img(img):
    plt.figure(figsize=(8,8))
    plt.axis('off')
    plt.imshow(img)
    plt.show()
```

```python
display_img(hsv)
```

```python
lower_red1 = np.array([0, 50, 50])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 50, 50])
upper_red2 = np.array([180, 255, 255])
```

```python
mask1 = cv2.inRange(hsv, lower_red1, upper_red1)
mask2 = cv2.inRange(hsv, lower_red2, upper_red2)
```

```python
mask = cv2.bitwise_or(mask1,mask2)
```

```python
mask = cv2.dilate(mask, np.ones((9,9), np.uint8), iterations=1)
mask = cv2.GaussianBlur(mask, (3,3), 0)
```

```python
display_img(mask)
```

```python
background = np.zeros(shape=img.shape)
```

```python
object_only = cv2.bitwise_and(img,img,mask=mask)
```

```python
display_img(object_only)
```

```python
hsv_object = cv2.cvtColor(object_only, cv2.COLOR_BGR2HSV)
```

```python
hsv_object[:,:,0][mask > 50] = 150
```

```python
final_object = cv2.cvtColor(hsv_object, cv2.COLOR_HSV2BGR)
```

```python
display_img(final_object)
```

```python
plt.figure(figsize=(10,8))
plt.title('All Black Background')
plt.axis('off')
plt.subplot(221)
plt.axis('off')
plt.title('Last Object')
plt.imshow(final_object)
plt.subplot(222)
plt.axis('off')
plt.title('Object with Mask Operations')
plt.imshow(mask)
plt.subplot(223)
plt.axis('off')
plt.title('Object HSV Color Space')
plt.imshow(hsv_object)
plt.subplot(224)
plt.axis('off')
plt.title('Default Object (BGR)')
plt.imshow(img)
```
